#### Q1. What is the mathematical formula for a linear SVM?



The mathematical formula for a linear Support Vector Machine (SVM) is used to define the decision boundary that separates two classes in a dataset. In a binary classification problem, a linear SVM seeks to find a hyperplane that maximizes the margin between the two classes. 
##### f(x) = w ⋅ x + b

- f(x): The decision boundary or decision function.
- x: The input data point you want to classify.
- w: The weight vector that is orthogonal (perpendicular) to the hyperplane.
- ⋅ : Dot product (also known as the inner product) between w and x.
- b: The bias term, also known as the intercept.

#### Q2. What is the objective function of a linear SVM?



The objective function of a linear Support Vector Machine (SVM) is designed to find the hyperplane that maximizes the margin between two classes in a binary classification problem. The SVM seeks to achieve this while minimizing the classification error.

Minimize:
#### 1/2∣∣w∣∣**2 + C Σmax(0,1−yi(w⋅xi+b))

- w: The weight vector that defines the hyperplane.
- C: A regularization parameter that controls the trade-off between maximizing the margin and minimizing the classification error. It's a hyperparameter set by the user, where a smaller value of C emphasizes a larger margin (potentially allowing some misclassification) and a larger value of C emphasizes correct classification (potentially leading to a smaller margin).

- Xi: The input data point.
- The class label associated with xi where yi can be either +1 or -1 for the two classes.
- b: The bias term (intercept) of the hyperplane.
- ∣∣w∣∣**2: The squared norm of the weight vector w. This term encourages the maximization of the margin.
- Σ: Summation over all data points in the training set.
- max(0,1−yi(w⋅xi+b)): This part of the objective function calculates the classification error or the margin violation for each data point. It's a hinge loss function that penalizes misclassification or points that fall inside the margin.

#### Q3. What is the kernel trick in SVM?



The kernel trick is a technique used in Support Vector Machines (SVMs) to extend their capabilities from linearly separable datasets to handle non-linearly separable data. It allows SVMs to implicitly transform the input data into a higher-dimensional feature space, where the data points may become linearly separable. This transformation is done using a mathematical function called a kernel function.

The primary idea behind the kernel trick is to avoid the explicit computation of the transformation into the higher-dimensional space, which would be computationally expensive or even impractical for high-dimensional spaces. Instead, the kernel function calculates the inner product (dot product) of the transformed data points directly in the higher-dimensional space, without explicitly computing the transformation itself.

#### 1. Linear Kernel: K(x, x') = x ⋅ x'
This is the same as the inner product in the original feature space and is used when the data is approximately linearly separable.

#### 2.   Polynomial Kernel: K(x, x') = (αx ⋅ x' + c)^d
This kernel introduces non-linearity by applying polynomial transformations to the data.

#### 3.  Radial Basis Function (RBF) or Gaussian Kernel: K(x, x') = exp(-γ||x - x'||^2)
The RBF kernel introduces a non-linear transformation based on the Gaussian distribution, which is effective for capturing complex patterns in the data.

#### 4.  Sigmoid Kernel: K(x, x') = tanh(αx ⋅ x' + c)
The sigmoid kernel introduces non-linearity using a hyperbolic tangent function.

#### Q4. What is the role of support vectors in SVM Explain with example




Support vectors play a crucial role in Support Vector Machines (SVMs) and are fundamental to the model's functionality. They are the data points that lie closest to the decision boundary (hyperplane) and have the maximum margin. Support vectors are used to define the decision boundary and are essential for the SVM's ability to generalize and make accurate predictions.

Consider a simple binary classification problem where you want to classify points in a 2D space into two classes, labeled as "A" and "B," using a linear SVM.

Example:

Class A: Red points
Class B: Blue points

In this example, the decision boundary is a straight line that separates the two classes. The support vectors are the data points from each class that are closest to the decision boundary or lie on the margin.

```python
       Margin
         |
         |
    B B  |           +     +
    B    |       +             +
    B    |   +                 +
    B    | +                   +
    B    +                     +
    B                          +   Support Vectors (Blue Points)
         +------------------------+
    A    |   +                 +
    A    |     +             +
    A    |       +         +
    A A  |           +     +
         |
         |
         +------------------------+
         |    A    |   B   |   A   |   B
```

In this example, the support vectors are the data points marked with "+" signs, both from class A and class B. These support vectors are critical because they determine the position of the decision boundary. The margin of the SVM is defined by the distance between the two parallel lines that touch these support vectors.

The SVM's objective is to maximize this margin while minimizing the classification error. By doing so, it aims to find the best decision boundary that separates the two classes while allowing for a certain level of misclassification (controlled by the regularization parameter C).

#### Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?



```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data
np.random.seed(42)
A = np.random.rand(20, 2) * 2 - 1  # Class A data points
B = np.random.rand(20, 2) * 2 + 1  # Class B data points
```

```python
plt.scatter(A[:, 0], A[:, 1], c='red', label='Class A')
plt.scatter(B[:, 0], B[:, 1], c='blue', label='Class B')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Scatter Plot of Data')
plt.grid(True)
plt.show()
```

Hyperplane

```python
plt.scatter(A[:, 0], A[:, 1], c='red', label='Class A')
plt.scatter(B[:, 0], B[:, 1], c='blue', label='Class B')
plt.axline((0, 0), slope=1, linestyle='--', color='green', label='Hyperplane')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Hyperplane')
plt.grid(True)
plt.show()
```

#### Marginal Plane : The marginal plane consists of two parallel lines that touch the support vectors

```python
plt.scatter(A[:, 0], A[:, 1], c='red', label='Class A')
plt.scatter(B[:, 0], B[:, 1], c='blue', label='Class B')
plt.axline((0, 0), slope=1, linestyle='--', color='green', label='Hyperplane')
plt.axline((0, 0), slope=1, linestyle='--', color='yellow', label='Marginal Plane')
plt.axline((0, 0), slope=1, linestyle='--', color='yellow')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Marginal Plane')
plt.grid(True)
plt.show()
```

- Hard Margin SVM does not allow any misclassification. It finds the maximum-margin hyperplane that perfectly separates the classes. If the data is not linearly separable, it fails.

- Soft Margin SVM allows for some misclassification to find a more robust solution. It introduces a regularization parameter C, which controls the trade-off between maximizing the margin and minimizing classification errors. A smaller C emphasizes a larger margin, potentially allowing some misclassification, while a larger C emphasizes correct classification, potentially leading to a smaller margin.

```python
from sklearn import svm

# Create a hard margin SVM
clf_hard = svm.SVC(kernel='linear', C=1e10)
clf_hard.fit(np.vstack((A, B)), np.array([1]*20 + [-1]*20))

# Create a soft margin SVM with a smaller C
clf_soft = svm.SVC(kernel='linear', C=0.1)
clf_soft.fit(np.vstack((A, B)), np.array([1]*20 + [-1]*20))

# Plot hard margin decision boundary
plt.scatter(A[:, 0], A[:, 1], c='red', label='Class A')
plt.scatter(B[:, 0], B[:, 1], c='blue', label='Class B')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = clf_hard.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Hard Margin SVM')
plt.grid(True)
plt.show()

# Plot soft margin decision boundary
plt.scatter(A[:, 0], A[:, 1], c='red', label='Class A')
plt.scatter(B[:, 0], B[:, 1], c='blue', label='Class B')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()
xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 50), np.linspace(ylim[0], ylim[1], 50))
Z = clf_soft.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Soft Margin SVM')
plt.grid(True)
plt.show()
```

#### Bonus task: Implement a linear SVM classifier from scratch using Python and compare its performance with the scikit-learn implementation.

```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Step 1: Load and prepare data
iris = datasets.load_iris()
X = iris.data[:, :2]  # We'll use the first two features for simplicity
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 2: SVM Training (Simplified)
def train_linear_svm(X, y, learning_rate=0.01, num_epochs=1000):
    # Initialize weights and bias
    w = np.zeros(X.shape[1])
    b = 0

    for epoch in range(num_epochs):
        for i, x_i in enumerate(X):
            # Calculate the margin
            margin = y[i] * (np.dot(w, x_i) + b)

            # Update weights and bias if margin is less than 1
            if margin < 1:
                w += learning_rate * (y[i] * x_i - 2 * w)
                b += learning_rate * y[i]

    return w, b

# Train the SVM
w, b = train_linear_svm(X_train, y_train)

# Step 3: SVM Prediction
def predict_linear_svm(X, w, b):
    return np.sign(np.dot(X, w) + b)

y_pred = predict_linear_svm(X_test, w, b)

# Step 4: Performance Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Simplified Linear SVM Accuracy: {accuracy:.2f}")

# Compare with scikit-learn SVM
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)
y_pred_sklearn = clf.predict(X_test)
accuracy_sklearn = accuracy_score(y_test, y_pred_sklearn)
print(f"Scikit-Learn Linear SVM Accuracy: {accuracy_sklearn:.2f}")
```
